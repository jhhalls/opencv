```python
# search the co-ordinates of the image
faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.05, minNeighbors=5)
faces2 = face_cascade.detectMultiScale(gray2, scaleFactor = 1.05, minNeighbors=5)
faces3 = face_cascade.detectMultiScale(gray3, scaleFactor = 1.05, minNeighbors=5)
faces4 = face_cascade.detectMultiScale(gray4, scaleFactor = 1.05, minNeighbors=5)
```

```python
faces5 = face_cascade.detectMultiScale(gray5, scaleFactor = 1.05, minNeighbors=5)
```

```python
print(type(faces))
print(faces)
```

```python
plt.imshow(faces, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
#  add rectangular face box around the face

for x,y,w,h in faces:
    image = cv2.rectangle(image, (x,y), (x+w, y+h), (0.,222,0), 3)

#  resie the image
resized = cv2.resize(image, (int(image.shape[1]), (int(image.shape[0]))))

#  show the image

plt.imshow(resized, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
